```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import copy
```

```python
source_dir = "images/"
save_dir = "save/"
```

```python
def show_image(image, title, save_name=None):
    fig = plt.figure(figsize=(10, 5))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.xticks([]), plt.yticks([])
    plt.title(title)
    if save_name is not None:
        plt.savefig(save_dir + save_name)
    plt.show()
```

```python
source_image = cv2.imread(source_dir + "Lines.jpg")
show_image(source_image, "original image")
```

```python
smoothed_image = cv2.bilateralFilter(source_image, 5, 120, 120)
# smoothed_image = cv2.GaussianBlur(source_image, (3, 3), 0)
show_image(smoothed_image, "Smoothed image")
```

```python
canny_edges = cv2.Canny(smoothed_image, 100, 250)
show_image(canny_edges, "Canny edges")
```

```python
lines = cv2.HoughLinesP(canny_edges, 1, np.pi/360, 80, minLineLength = 5, maxLineGap = 10)
image_with_lines = copy.deepcopy(source_image)

for x0, y0, x1, y1 in lines[:, 0, :]:
    cv2.line(image_with_lines, (x0, y0), (x1, y1), (255, 0, 0), 2)

show_image(image_with_lines, "detected_lines", "1-detected_lines")
```
