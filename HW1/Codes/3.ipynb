```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import copy
```

```python
source_dir = "images/"
save_dir = "save/"
```

```python
image = cv2.imread(source_dir + "Blur.png", 0)
```

```python
A_list = [3, 6, 10, 20, 50]

plt.figure(figsize=(15, 15))
plt.subplot(321)
plt.imshow(image, cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("Original image")

for i, A in enumerate(A_list):
    kernel = [[0, -A, 0],
             [-A, 4*A + 1, -A],
             [0, -A, 0]]

    kernel = np.asarray(kernel)
    
    sharpened_image = cv2.filter2D(image, -1, kernel)
    plt.subplot(3, 2, i+2)
    plt.imshow(sharpened_image, cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("Sharpened image with A = %d"%(A))

plt.savefig(save_dir + "sharpened_image")
plt.show()
```
