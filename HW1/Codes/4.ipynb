```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import copy
```

```python
source_dir = "images/"
save_dir = "save/"
```

```python
image = cv2.imread(source_dir + "Edge.png", 0)
smoothed_image = cv2.GaussianBlur(image, (3, 3), 0)
```

### LoG operator with different sizes

```python
kernel_size_list = [3, 5, 7]
depth = cv2.CV_64F

plt.figure(figsize=[15, 10])
plt.subplot(221)
plt.title("Original_image")
plt.imshow(image, cmap="gray"), plt.xticks([]), plt.yticks([])

for i, kernel_size in enumerate(kernel_size_list):
    smoothed_log = cv2.Laplacian(smoothed_image, depth, ksize = kernel_size)
    smoothed_log = cv2.convertScaleAbs(smoothed_log)

    plt.subplot(2, 2, i+2)
    plt.title("LoG edge detection with kernel_size = %d"%kernel_size)
    plt.imshow(smoothed_log, cmap="gray"), plt.xticks([]), plt.yticks([])

plt.savefig(save_dir + "LoG")
plt.show()
```

```python
kernel_size_list = [3, 5, 7]
depth = cv2.CV_64F

plt.figure(figsize=[15, 10])
plt.subplot(221)
plt.title("Original_image")
plt.imshow(image, cmap="gray"), plt.xticks([]), plt.yticks([])

for i, kernel_size in enumerate(kernel_size_list):
    smoothed_sobel = cv2.Sobel(smoothed_image, depth, 1, 1, ksize = kernel_size)
    smoothed_sobel = cv2.convertScaleAbs(smoothed_sobel)
    
    plt.subplot(2, 2, i+2)
    plt.title("Sobel edge detection with kernel_size = %d"%kernel_size)
    plt.imshow(smoothed_sobel, cmap="gray"), plt.xticks([]), plt.yticks([])

plt.savefig(save_dir + "Sobel")
plt.show()
```

```python
# smoothed_image = cv2.GaussianBlur(image, (5, 5), 0)
tuple_list = [(50, 200),
              (100, 200),
              (100, 250)]

plt.figure(figsize=[15, 10])
plt.subplot(221)
plt.title("Original_image")
plt.imshow(smoothed_image, cmap="gray"), plt.xticks([]), plt.yticks([])

for i, (low, high) in enumerate(tuple_list):
    smoothed_canny = cv2.Canny(smoothed_image, low, high)
    
    plt.subplot(2, 2, i+2)
    plt.title("Canny edge detection with thresholds = (%d, %d)"%(low, high))
    plt.imshow(smoothed_canny, cmap="gray"), plt.xticks([]), plt.yticks([])

plt.savefig(save_dir + "Canny")
plt.show()
```
