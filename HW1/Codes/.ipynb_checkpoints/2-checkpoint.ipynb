```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import copy
```

```python
source_dir = "images/"
save_dir = "save/"
```

```python
image = cv2.imread(source_dir+"Noise.png", 0) / 255
plt.imshow(image, cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("The original image")
plt.show()
```

### Adding the guassian Noise

```python
shape = image.shape
mean = 0
variance = 0.2
noise = np.random.normal(mean, variance, shape)
noisy_image = image + noise

cv2.imwrite(save_dir + "/normal.jpg", noisy_image * 255)

plt.figure(figsize=[15, 8])
plt.subplot(121)
plt.title("Noisy image")
plt.imshow(noisy_image, cmap="gray"), plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.title("Noise")
plt.imshow(noise, cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

### Salt and Pepper noise

```python
noisy_image = np.copy(image)
salt_to_pepper = 0.5
intensity = 0.1
size = image.size

# Salt mode
pixels = [np.random.randint(0, i - 1, int(size * intensity * salt_to_pepper))
         for i in image.shape]

noisy_image[pixels] = 1

# Pepper mode
pixels = [np.random.randint(0, i - 1, int(size * intensity * (1. - salt_to_pepper)))
         for i in image.shape]
noisy_image[pixels] = 0

cv2.imwrite(save_dir + "/s&p.jpg", noisy_image * 255)

plt.title("Image with Salt and Pepper noise")
plt.imshow(noisy_image, cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

### Uniform Noise (Quantization noise) To be checked later

```python
shape = image.shape
noise = np.random.uniform(-0.2, 0.2, shape)
noisy_image = image + noise

cv2.imwrite(save_dir + "/uniform.jpg", noisy_image * 255)

plt.title("Image with Uniform noise")
plt.imshow(noisy_image, cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=[15, 5])
plt.subplot(121)
plt.title("Original histogram")
plt.hist((image * 255).ravel(), 256, [0,256])

plt.subplot(122)
plt.title("noisy image histogram")
plt.hist((noisy_image * 255).flatten(), 256, [0, 256])
plt.show()
```

### Adding poisson noise (shot noise)

```python
shape = image.shape
lam = 0.1
noise = np.random.poisson(lam, shape)
noisy_image = image + noise
plt.show()

cv2.imwrite(save_dir + "/shot.jpg", noisy_image * 255)

plt.title("Image with Uniform noise")
plt.imshow(noisy_image, cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=[15, 5])
plt.subplot(121)
plt.title("Original histogram")
plt.hist((image * 255).ravel(), 256, [0,256])

plt.subplot(122)
plt.title("noisy image histogram")
plt.hist((noisy_image * 255).flatten(), 256, [0, 256])
plt.show()
```

## Denoising the images

```python
def show_plots(noisy_image, noise_type):
    plt.figure(figsize=(15, 15))
    plt.subplot(321)
    plt.suptitle("Considering the results of filters on %s noise type"%(noise_type), size=20)
    plt.imshow(image, cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("original image")
    plt.subplot(322)
    plt.imshow(noisy_image, cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("image with noise")
    plt.subplot(323)
    kernel55 = np.ones((5, 5)) * (1 / 25)
    plt.imshow(cv2.filter2D(noisy_image, -1, kernel55), cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("average smoothing")
    plt.subplot(324)
    plt.imshow(cv2.medianBlur(noisy_image, 5), cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("median smoothing")
    plt.subplot(325)
    plt.imshow(cv2.GaussianBlur(noisy_image, (5, 5), 0), cmap="gray"), plt.xticks([]), plt.yticks([]), plt.title("guassian smoothing")
    plt.savefig(save_dir + noise_type +"-denoising.jpg")
    plt.show()
```

```python
normal_noise = cv2.imread(save_dir+"normal.jpg", 0)
show_plots(normal_noise, "guassian")
```

```python
normal_noise = cv2.imread(save_dir+"s&p.jpg", 0)
show_plots(normal_noise, "salt and pepper")
```

```python
normal_noise = cv2.imread(save_dir+"shot.jpg", 0)
show_plots(normal_noise, "shot")
```

```python
normal_noise = cv2.imread(save_dir+"uniform.jpg", 0)
show_plots(normal_noise, "uniform")
```
