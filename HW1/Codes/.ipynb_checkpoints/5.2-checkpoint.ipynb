```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import copy
```

```python
source_dir = "images/"
save_dir = "save/"
```

```python
image = cv2.imread(source_dir + "Objects2.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

### Canny edge detection

```python
smoothed_image = cv2.GaussianBlur(image, (15, 15), 0)
edges = cv2.Canny(smoothed_image, 50, 230)
plt.imshow(edges, cmap="gray")

plt.savefig(save_dir + "11-edges")
plt.show()
```

```python
plt.figure(figsize=(15, 5))

x_hist = np.sum(np.where(edges != 255, edges, 1), axis=0)
y_hist = np.sum(np.where(edges != 255, edges, 1), axis=1)

plt.subplot(121)
plt.grid(alpha=0.5)
plt.bar(list(range(edges.shape[1])), x_hist, color="blue"), plt.title("X histogram")
plt.subplot(122)
plt.grid(alpha=0.5)
plt.bar(list(range(edges.shape[0])), y_hist, color="blue"), plt.title("Y histogram")

plt.savefig(save_dir + "12-initial_histogram")
plt.show()
```

#### Using the above histograms to crop the original image and restrict the X hostogram to the plate itself

```python
cropped_image = image[np.where(y_hist > 0)[0][0]:np.where(y_hist > 0)[0][-1],
                      np.where(x_hist > 0)[0][0]:np.where(x_hist > 0)[0][-1]]
plt.imshow(cropped_image, cmap="gray")
plt.savefig(save_dir + "13-cropped_plate")
plt.show()
```

```python
smoothed_image_cropped = cv2.GaussianBlur(cropped_image, (13, 13), 0)
edges_cropped = cv2.Canny(smoothed_image_cropped, 210, 250)
plt.imshow(edges_cropped, cmap="gray")
plt.savefig(save_dir + "14-canny_edges")
plt.show()
```

```python
plt.figure(figsize=(15, 5))

x_hist = np.sum(np.where(edges_cropped != 255, edges_cropped, 1), axis=0)
y_hist = np.sum(np.where(edges_cropped != 255, edges_cropped, 1), axis=1)

plt.subplot(121)
plt.grid(alpha=0.5)
sns.barplot(list(range(edges_cropped.shape[1])), x_hist, color="blue"), plt.title("X histogram")
plt.subplot(122)
plt.grid(alpha=0.5)
sns.barplot(list(range(edges_cropped.shape[0])), y_hist, color="blue"), plt.title("Y histogram")
plt.savefig(save_dir + "15-secondary histograms")
plt.show()
```

### Defining the boundries
In order to make the process prone to noise, we calculate the moving average of three consecutive values.

```python
window_size = 9
x_hist_list = list(x_hist)

# Padding the list of the histogram to maintain its shape
x_hist_list = [0] * (window_size // 2) + x_hist_list + ([0] * (window_size // 2))

# Smoothing the histogram with a 3 windows size
averaged_list = []
for i in range(len(x_hist_list) - window_size + 1):
    averaged_list.append(sum(x_hist_list[i: i + window_size]) / window_size)
averaged_list_x = np.asarray(averaged_list)

# Plotting the smoothed histogram
assert averaged_list.__len__() == x_hist.__len__()
plt.figure(figsize=(20, 5))
plt.grid(alpha=0.5)
plt.bar(list(range(edges_cropped.shape[1])), averaged_list_x, color="blue"), plt.title("X histogram")
plt.xticks(list(range(0, edges_cropped.shape[1], 20)))

plt.savefig(save_dir + "16-smoothed histogram")
plt.show()
```

```python
window_size = 9
y_hist_list = list(y_hist)

# Padding the list of the histogram to maintain its shape
y_hist_list = [0] * (window_size // 2) + y_hist_list + ([0] * (window_size // 2))

# Smoothing the histogram with a 3 windows size
averaged_list = []
for i in range(len(y_hist_list) - window_size + 1):
    averaged_list.append(sum(y_hist_list[i: i + window_size]) / window_size)
averaged_list_y = np.asarray(averaged_list)

# Plotting the smoothed histogram
assert averaged_list.__len__() == y_hist.__len__()
plt.figure(figsize=(20, 5))
plt.grid(alpha=0.5)
plt.bar(list(range(edges_cropped.shape[0])), averaged_list_y, color="blue"), plt.title("Y histogram")
plt.xticks(list(range(0, edges_cropped.shape[1], 20)))

plt.savefig(save_dir + "16-smoothed histogram")
plt.show()
```

```python
def find_boundries(histogram, threshold=7, counter_threshold=2):
    in_boundry_region = False
    counter = 0
    boundries = [0]
    for i, value in enumerate(histogram):
        
        if in_boundry_region:
            counter += 1
        
        if value <= threshold and not in_boundry_region:
            begin_boundry = i
            in_boundry_region = True

        if value > threshold and counter >= counter_threshold:
            boundries.append((begin_boundry + i) // 2)
            in_boundry_region == False
            counter = 0
            
        if value > threshold and counter <= counter_threshold:
            counter = 0
            in_boundry_region = False

        # Appending the final boundry
    boundries.append(i)
    return boundries
```

```python
boundries_x = find_boundries(averaged_list_x)
boundries_y = find_boundries(averaged_list_y)

# for i in range(averaged_list_x.shape[0]):
#     print(averaged_list_x[i], averaged_list_y[i])

shape = [5, 5]
for i, (start_x, end_x) in enumerate(zip(boundries_x[:-1], boundries_x[1:])):
    for j, (start_y, end_y) in enumerate(zip(boundries_y[:-1], boundries_y[1:])):
        plt.subplot(shape[0], shape[1], i*shape[0] + j+1)
        plt.imshow(cropped_image[start_y:end_y, start_x:end_x]), plt.xticks([]), plt.yticks([])

plt.savefig(save_dir + "17-cropped_images")
plt.show()
```
