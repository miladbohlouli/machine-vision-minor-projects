```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import os
from time import time

source_dir = ["Datasets/" + ds + "/" for ds in os.listdir("Datasets")[:-1]]
```

```python
def compare(*images, **args):
    imgL = cv.imread(images[0], 0)
    imgR = cv.imread(images[1], 0)
    
    stereo = cv.StereoBM_create(**args)
    disparity = stereo.compute(imgL,imgR)
    plt.imshow(disparity,'gray'), plt.xticks([]), plt.yticks([])
```

```python
first_images_set = [source_dir[0] + str for str in sorted(os.listdir(source_dir[0]))]
first_image_pairs = [(images[0], images[1]) for images in zip(first_images_set[2:-1], first_images_set[3:])]

plt.figure(figsize=[8, 8])
for i in range(len(first_image_pairs)):
    plt.subplot(4, 3, i+1)
    name1, name2 = first_image_pairs[i][0].split(".")[0].split("/")[-1], first_image_pairs[i][1].split(".")[0].split("/")[-1]
    plt.title(" and ".join([name1, name2]))
    compare(*first_image_pairs[i], numDisparities=16, blockSize=15)
save_fig()
```

```python
second_images_set = [source_dir[1] + str for str in sorted(os.listdir(source_dir[1]))]
second_image_pairs = [(images[0], images[1]) for images in zip(second_images_set[:-2], second_images_set[1:-1])]

plt.figure(figsize=[10, 10])
for i in range(len(second_image_pairs)):
    plt.subplot(2,2,i+1)
    name1, name2 = second_image_pairs[i][0].split(".")[-2], second_image_pairs[i][1].split(".")[-2]
    plt.title(" and ".join([name1, name2]))
    compare(*second_image_pairs[i], numDisparities=16, blockSize=15)
save_fig()
```
